```python
import pandas as pd
import matplotlib.pyplot as plt
mechanics_df = pd.read_csv('10 filters to 1 filter.csv')
mechanics_df = mechanics_df[mechanics_df['Time (s) Run #1'].notna()]
for x in mechanics_df.index:
    if mechanics_df.loc[x, 'Time (s) Run #1'] > 1.86 or mechanics_df.loc[x, 'Time (s) Run #1'] < 1:
        mechanics_df.drop(x, inplace = True)
mechanics_df.plot(kind = 'line', x = 'Time (s) Run #1', y = 'Position (m) Run #1', title = 'Position vs Time graph')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
mechanics_df = pd.read_csv('10 filters to 1 filter.csv')
mechanics_df = mechanics_df[mechanics_df['Velocity (m/s) Run #1'].notna()]
for x in mechanics_df.index:
    if mechanics_df.loc[x, 'Velocity (m/s) Run #1'] > 0 or mechanics_df.loc[x, 'Time (s) Run #1'] > 1.86:
        mechanics_df.drop(x, inplace = True)
mechanics_df.plot(kind = 'line', x = 'Time (s) Run #1', y = 'Velocity (m/s) Run #1', title = 'Velocity vs Time graph')
```
